```python
import yfinance as yf
import pandas as pd

# Download Tesla stock data
tesla_data = yf.download('TSLA')

# Reset the index
tesla_data.reset_index(inplace=True)

# Save the DataFrame as a CSV file
tesla_data.to_csv("tesla_stock_data.csv", index=False)

# Display the first five rows
tesla_data.head()
```

```python
import yfinance as yf
import pandas as pd

# Get Tesla stock ticker data
tesla = yf.Ticker("TSLA")

# Get Tesla revenue data
tesla_revenue = tesla.financials.loc["Total Revenue"]

# Convert revenue data into a DataFrame
tesla_revenue_df = pd.DataFrame(tesla_revenue).reset_index()
tesla_revenue_df.columns = ["Date", "Revenue"]

# Save the data as a CSV file
tesla_revenue_df.to_csv("tesla_revenue_data.csv", index=False)

# Display the last five rows
tesla_revenue_df.tail()
```

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Download Tesla stock data
tesla_data = yf.download('TSLA')

# Reset the index
tesla_data.reset_index(inplace=True)

# Function to plot Tesla stock graph
def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.xticks(rotation=45)
    plt.show()

# Plot the graph with a title
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
